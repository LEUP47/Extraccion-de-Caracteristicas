```python
import funpymodeling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from funpymodeling.exploratory import freq_tbl
```

```python
mr2=pd.read_csv("mr2.csv")
```

```python
#obtengo un analisis de las variables categoricas
freq_tbl(mr2)
```

```python
table1=freq_tbl(mr2['210_sales_channels'])
table1
```

```python
filtro1=table1[table1['frequency']>7]
filtro1
```

```python
filtro_index=filtro1.set_index('210_sales_channels')
filtro_index
```

```python
#Realizamos grafico de pastel del dataframe filtrado
filtro_index["frequency"].plot(kind='pie', figsize=(10,5), shadow=True, autopct="%0.1f%%")
```
