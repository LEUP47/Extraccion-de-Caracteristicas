```python
import funpymodeling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from funpymodeling.exploratory import freq_tbl
```

```python
mr2=pd.read_csv("mr2.csv")
```

```python
table1=freq_tbl(mr2['185_place_orders_suppliers'])
table1
```

```python
filtro1=table1[table1['frequency']>10]
filtro1
```

```python
filtro_index=filtro1.set_index('185_place_orders_suppliers')
filtro_index
```

```python
filtro_index_frequency=filtro_index.drop(['percentage','cumulative_perc'],axis=1)
filtro_index_frequency
```

```python
filtro_index_frequency.plot(kind='bar',width=0.8,figsize=(12,10))
plt.title('Contacto con proveedores')
plt.xlabel('Medio de contacto')
plt.ylabel("Frecuencia")
```
