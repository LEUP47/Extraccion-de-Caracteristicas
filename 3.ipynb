```python
import funpymodeling
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from funpymodeling.exploratory import freq_tbl
```

```python
mr2=pd.read_csv("mr2.csv")
```

```python
mr2.head()
```

```python
table1=freq_tbl(mr2['19_tax_id'])
table1
```

```python
filtro1=table1[table1['frequency']>5]
filtro1
```

```python
filtro_index=filtro1.set_index('19_tax_id')
filtro_index
```

```python
filtro_index_frequency=filtro_index.drop(['percentage','cumulative_perc'],axis=1)
filtro_index_frequency
```

```python
filtro_index_frequency.plot(kind='bar',width=0.8,figsize=(2,1))
plt.title('19_tax_id')
plt.xlabel('Tax id')
plt.ylabel("Frecuencia")
```
